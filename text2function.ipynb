# Semantic Search

```python
def semantic_search(query:str, documents:typing.List[str], number_of_matches = 5):
    """Search a list of _documents_ against a query
    """

    query_embedding = embedder.encode(query, convert_to_tensor=True)
    document_embeddings = embedder.encode(documents, convert_to_tensor=True)
    cos_scores = sentence_transformers.util.cos_sim(query_embedding, document_embeddings)[0]
    top_matches = torch.topk(cos_scores, k=number_of_matches)
    
    top_matching_documents = [(documents[idx], score) for score, idx in zip(top_matches[0], top_matches[1]) ]

    for document, score in top_matching_documents:
        print (document.ljust(80, "-"), " {:.2f}% match".format(score*100))

    return top_matching_documents
```

# Queries (Numpy functions)

```python
import numpy
import numpy.matlib
from numpydoc.docscrape import NumpyDocString

file_data = open('aaaaa - aaaa.csv')
functions = []
for row in file_data:
    functions.append(eval(row.replace("\n", "")))
```

```python
queries = []

for function in functions:
  temp = []
  doc = NumpyDocString(function.__doc__)
  
  temp_summary = doc["Summary"]

  summary = ''
  for i in temp_summary:
    summary = summary + " " + i

  queries.append([function.__name__, summary])
```

```python
browser.get('https://stackoverflow.com/questions/tagged/numpy?sort=votes&page=1&pagesize=50')

soup = BeautifulSoup(browser.page_source, 'html.parser')

questions1=soup.find_all('a', class_='s-link')
```

```python
import re

list_questions = []

for question in questions1:
  
  final_question = ""
  for index in re.finditer('/', question['href']):
          final_question = question['href'][index.end():]

  final_question = final_question.replace('-',' ')
  list_questions.append(final_question)


for question in questions2:
  
  final_question = ""
  for index in re.finditer('/', question['href']):
          final_question = question['href'][index.end():]

  final_question = final_question.replace('-',' ')
  list_questions.append(final_question)
```

```python
result = []
i = 0

for query in queries:
  temp_result = semantic_search(query[1], list_questions, number_of_matches=len(list_questions))
  
  result.append([i, query[0], temp_result[0:5]])
  i += 1
```
